```python
# Create a class
class Add:
    def __init__(self, num_1, num_2):
        self.num_1 = num_1
        self.num_2 = num_2

    def sum_all(self):
        print("Method sum_all in class Add")
        return self.num_1 + self.num_2


# Crease an instance of Add
a=Add(30,10)

# Now we can access the function within the class
# using the dot notation
display(a.sum_all())
```
